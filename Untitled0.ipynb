```python
def get_string(text):
    if pd.isna(text):
        return text
    
    text = re.sub("[\d\.'\-(nan)\[\]\/]", "", text)
    text = re.sub(" ", "", text)
    text = re.sub(",{2,}", ",", text)
    return text
```

```python
get_string("[1. '화장실' 5, 2. 주차장, -, nan, /]")
```

```python
get_string(np.nan)
```

```python
df["편익시설"] = df["공원보유시설(편익시설)"].map(get_string)
df["편익시설"].head()
```

```python
import folium

m = folium.Map([35.900539, 128.604905], zoom_start=13)


for n in df_sn.index:
    lat = df_sn.loc[n, "위도"]
    long = df_sn.loc[n, "경도"]
    tooltip = df_sn.loc[n, "공원명"] + " " + df_sn.loc[n, "소재지도로명주소"]
    
    folium.Marker([lat, long], tooltip=tooltip).add_to(m)

m.save('index.html')
m
```
